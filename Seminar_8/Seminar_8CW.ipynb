Уровень статистической значимости принять за 5%
1 . Используйте функцию в Python:
2. Имея p-value из функции рассчитать наблюдаемое значение критерия.


```python
a = np.array([12, 10, 11, 19, 13, 11, 17, 15, 19, 14, 21, 18, 21, 11, 17, 14, 15, 17, 20, 19])
b = np.array([11, 13, 18, 15, 17, 18, 10, 21, 26, 15, 11, 12, 15, 17, 10, 18, 18, 12,21, 20])

std_a = np.std(a, ddof=1) # с находим среднее квадротичное отклонение первой выборки (ddof=1, где единица степень свободы)
std_b = np.std(b, ddof=1) # Среднее квадротичное отклонение второй выборки

len(a), len(b) # Кол-во выборок 

d = ((std_a ** 2 + std_b ** 2) / 20) ** 0.5 # Находим общую дисперсию 
```

```python
(np.mean(a) - np.mean(b)) / d # находим значение статистики
```

```python
import scipy.stats as stats

stats.ttest_ind(a, b, equal_var=True) # Проводим двухсторонний тест
```
